
## âœ´ <u>Handling the NaN/null/missing values in each attribute of the dataframe.</u>

<br>

> let's create a nice little dataframe which contain the metadata.

```python
# let's create a custom function to create a dataframe which contain the metadata about main dataframe.
# define a function.
def metainfo():
  
  # doc string contain a short description about the function useability.
  '''This function return the new dataframe called meta_structure, & gives metadata about dataframe.'''
  
  # we want each attribute of the dataframe as a row in new datafrme.
  meta_structure = pd.DataFrame(index=ps_df.columns)

  # let's create a six new column which defines our metadata.
  meta_structure["data_type"] = ps_df.dtypes
  meta_structure["not_null_values"] = ps_df.count()
  meta_structure["null_values"] = ps_df.isnull().sum()
  meta_structure["null_values_percentage"] = round(ps_df.isnull().mean(),5) * 100
  meta_structure["unique_count"] = ps_df.nunique()
  meta_structure["duplicated_count"] = ps_df.shape[0] - ps_df.nunique()

  # just return the created structure.
  return meta_structure
```

```python
# function calling.
meta_df = metainfo()

# let's print the meta_df
meta_df
```

### âš™ <b>4. <code>Rating</code>: Attribute contain 1470 null value.</b>

```python
# locating the null values in the Rating attribute.
ps_df[ps_df['Rating'].isnull()]
```

1. As we know that the null values is account to approx 13.6% of the total records.
2. So, dropping the records is not a solution because, it may case a loss of data and significatly altered our analysis.
3. Also, dropping the data is always the costly operation if it's account to a significant percentage of the total records.
4. For this case we have to adopt the different solution.

<p>First we have to check if any value in the <code>Rating</code> attribute is outside of the range 1 to 5. As we know that the rating of the app is out of 5 only.</p>

```python
# checking the range of rating.
ps_df[(ps_df['Rating'] <1) | (ps_df['Rating']>5)]
```

<p><b>From the above we conclude that the all values are well with a range.</b></p>

```python
# Finding mean in the Rating column excluding the NaN values.
mean_rating = round(ps_df[~ps_df['Rating'].isnull()]['Rating'].mean(),2)

# Finding median in the Rating column excluding the NaN values.
median_rating = ps_df[~ps_df['Rating'].isnull()]['Rating'].median()

# output the mean and median
[mean_rating , median_rating]
```


<p><b>ðŸŸ¨ The biggest question is which value to take mean or median.â“</b></p> 

<br>

1. let's take some help of visulization to answer this burning question.
2. To see the distribution of rating we use displot.
3. To see the outliers we use boxplot.

```python
# setting-up the sizes
fig, ax = plt.subplots(2,1, figsize=(20,10))

# plotting the distplot
sns.distplot(ps_df['Rating'],color='firebrick',ax=ax[0])

# plotting the boxplot
sns.boxplot(x='Rating',data=ps_df, ax=ax[1])
```

ðŸŸ¨ Conclusions : 

1. The mean of the average ratings (excluding the NaN values) comes to be 4.2.

2. The median of the entries (excluding the NaN values) in the 'Rating' column comes to be 4.3. 

3. From this we can say that 50% of the apps have an average rating of above 4.3, and the rest below 4.3.

4. From the distplot visualizations, it is clear that the ratings are left skewed.

5. We know that if the variable is skewed, the mean is biased by the values at the far end of the distribution. Therefore, the median is a better representation of the majority of the values in the variable.

6. Hence we will impute the NaN values in the Rating column with its median.

```python
# Replacing the NaN values in the 'Rating' colunm with its median value
ps_df['Rating'].fillna(value=median_rating,inplace=True)
```

```python
# verifying the changes by calling the metainfo function.
metainfo()
```

<p>We are clearly seeing that the all values belong the attribute <code>null_values</code> are come down to zero.</p>
