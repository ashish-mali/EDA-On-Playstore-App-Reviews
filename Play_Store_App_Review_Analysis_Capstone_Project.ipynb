## ‚ú¥ <u>What is Exploratory Data Analysis?</u>

> <b>Defination:<br> Exploratory Data Analysis refers to the critical process of <u>performing initial investigations on data so as to discover patterns,
</u>to spot anomalies,to test hypothesis and to check assumptions with the help of summary statistics and graphical representations.</b>
or<br>
<b>Exploratory Data Analysis (EDA) is an approach to analyze the data using visual techniques. It is used to discover trends, patterns, or to check assumptions with the help of statistical summary and graphical representations. </b><br>

Data scientists use exploratory data analysis (EDA) to examine data sets for patterns and abnormalities (outliers), formulate hypotheses based on the understanding of the information, and describe their key characteristics based on the domain knowledge. Visual analytics techniques are frequently employed in exploratory data analysis. In any project involving analysis of the data or in overall data science, it is a crucial stage. It assists in determining how to best manipulate data sources to achieve the desired outcomes. <br>

In order to better comprehend the data and make it easy to understand and get the quick view of the large numerical data, EDA generally employes the various charts and graphics to better understand the data in the summery format.
