#### ↪ <b>The Scales Of Data Measurement</b>
<br>

> It is generally understood that a specific characteristic (feature / column) of structured data can be broken down into aone of four levels of data. the levels are :

>*   The nominal level
>*   The ordinal level
>*   The interval level
>*   The ratio level.

> As we move down the list we gain more structure and therefore more returns from our analysis. Each level comes with it's own accepted practicve in measuring the <code>conter</code> of the data. We usually think of the mean/average as being an acceptable form of center, howerver, this is only true for a specific type of data. 

> let's brifly define each of these levels.
> 1. <b>The nominal level</b> :- A nominal scale describes a variable with categories that do not have a natural order or ranking. You can code nominal variables with numbers if you want, but the order is arbitrary and any calculations, such as computing a mean, median, or standard deviation, would be meaningless.<br>
> 2. <b>The ordinal level</b> :- An ordinal scale is one where the order matters but not the difference between values.<br>
> 3. <b>The interval level</b> :- An interval scale is one where there is order and the difference between two values is meaningful.<br>
> 4. <b>The ratio level</b> :- An extension of the interval level of measurement. It deals with data that have a natural zero point. The difference between the values, and the ratio of values, both are meaningful in the ratio level of measurement.

```python
# let's create a custom function to create a dataframe which contain the metadata about main dataframe.
# define a function.
def metainfo(given_df):
  
  # doc string contain a short description about the function useability.
  '''This function return the new dataframe called meta_structure, & gives metadata about dataframe.'''
  
  # we want each attribute of the dataframe as a row in new datafrme.
  meta_structure = pd.DataFrame(index=given_df.columns)

  # let's create a six new column which defines our metadata.
  meta_structure["data_type"] = given_df.dtypes
  meta_structure["not_null_values"] = given_df.count()
  meta_structure["null_values"] = given_df.isnull().sum()
  meta_structure["null_values_percentage"] = round(given_df.isnull().mean(),5) * 100
  meta_structure["unique_count"] = given_df.nunique()
  meta_structure["duplicated_count"] = given_df.shape[0] - given_df.nunique()

  # just return the created structure.
  return meta_structure
```

> 1. First of all, we see that the <code>Size</code> attribute contains mainly two distinct unit of measurement for the size of the application such as Megabits(MB) and Kilobits(KB). So, for easy analysis we must have to convert these into a common unit of measurement. we choose MB for this size representation.<br>
> 2. We all know that, 1MB=1024KB, now to convert all KB's to MB, we must divide each KB by 1024.

```python
# Defining a function to convert all the entries in KB to MB and then converting them to float datatype.
# defining a function
def convert_kb_to_mb(row):
  # doc string for function information.
  '''This function converts all the valid entries in KB to MB and returns the result in float datatype.'''
  try:
    # if entry is in MB represented as 'M ' in attribute.
    if 'M' in row:
      # remove suffix and convert rest in float data type.
      return float(row[:-1])
    # if entry is in KB represented as 'K' in attribute.
    elif 'k' in row:
       # remove suffix and convert rest in float data type.
      return round(float(row[:-1])/1024, 3)
    # if encountered with any other case just return that record.
    else:
      return row
  # if anywhere in the execution case an exception handling it below.
  except:
    return row
```

```python
# let's apply the above function to each row of the 'Size' attribute.
ps_df['Size'] = ps_df['Size'].apply(lambda x: convert_kb_to_mb(x))

# review the transformation.
ps_df
```
